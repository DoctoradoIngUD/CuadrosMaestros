```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
Articulos = pd.read_csv('articulos.csv', sep=";")
Cuartiles = pd.read_csv('scimagojr.csv', sep=";")
Cuartiles=Cuartiles.drop(['Rank','Title','Sourceid','Type','SJR','H index','Total Docs. (2022)','Total Docs. (3years)','Total Refs.','Total Cites (3years)','Citable Docs. (3years)',
                          'Cites / Doc. (2years)','Ref. / Doc.','Country','Region','Publisher','Coverage','Categories','Areas'], axis=1)
Articulos  = Articulos.replace({'                ISSN': '-'},{'                ISSN': ''}, regex=True)
Articulos  = Articulos.replace({'                ISSN': ' '},{'                ISSN': ''}, regex=True)
#Cuantiles
Cuartil = pd.DataFrame(columns=['Cuartil'],index = range(Articulos.shape[0]))
Cuartil = Cuartil.fillna('No Clasificado')
Articulos = pd.concat([Articulos,Cuartil],axis=1)
# Publindex
Publindex = pd.DataFrame(columns=['Publindex'],index = range(Articulos.shape[0]))
Publindex = Publindex.fillna('No Clasificado')
Articulos = pd.concat([Articulos,Publindex],axis=1)
```

```python
for i in range(len(Articulos['                ISSN'])):
    for j in range(len(Cuartiles['Issn'])):
        if Articulos['                ISSN'][i] in Cuartiles['Issn'][j]:
            Articulos['Cuartil'][i] = Cuartiles['SJR Best Quartile'][j]
```

```python
publindex = pd.read_csv('publindex.csv', sep=",")
publindex  = publindex.replace({'TXT_ISSN_P': '-'},{'TXT_ISSN_P': ''}, regex=True)
publindex  = publindex.replace({'TXT_ISSN_P': ' '},{'TXT_ISSN_P': ''}, regex=True)
publindex['TXT_ISSN_P'] = publindex['TXT_ISSN_P'].astype(str)
```

```python
for l in range(len(Articulos['                ISSN'])):
    for k in range(len(publindex['TXT_ISSN_P'])):
        if Articulos['                ISSN'][l] in publindex['TXT_ISSN_P'][k]:
            Articulos['Publindex'][l] = publindex['ID_CLAS_REV'][k]
```

```python
Articulos.to_csv('articss.csv')
```

```python
Conteo = pd.read_csv('articss.csv', sep=",")
```

```python

Conteo.drop(Conteo[(Conteo['                AÑO'] < 2016)].index, inplace=True)
####
##Articulos Q1
Nombre = Conteo[:]
Cuartil1 = Conteo[:]
Cuartil2 = Conteo[:]
Cuartil3 = Conteo[:]
Cuartil4 = Conteo[:]
PIndexA1 = Conteo[:]
PIndexA2 = Conteo[:]
PIndexB = Conteo[:]
PIndexC = Conteo[:]
```

```python
Nombre = Nombre.drop(['                AÑO','                LUGAR','                REVISTA','                EDITORIAL','                ISSN','Unnamed: 8','Cuartil','Publindex'], axis=1)
Cuartil1.drop(Cuartil1[(Cuartil1['Cuartil'] != "Q1")].index, inplace=True)
Cuartil2.drop(Cuartil2[(Cuartil2['Cuartil'] != "Q2")].index, inplace=True)
Cuartil3.drop(Cuartil3[(Cuartil3['Cuartil'] != "Q3")].index, inplace=True)
Cuartil4.drop(Cuartil4[(Cuartil4['Cuartil'] != "Q4")].index, inplace=True)
PIndexA1.drop(PIndexA1[(PIndexA1['Publindex'] != "A1")].index, inplace=True)
PIndexA2.drop(PIndexA2[(PIndexA2['Publindex'] != "A2")].index, inplace=True)
PIndexB.drop(PIndexB[(PIndexB['Publindex'] != "B")].index, inplace=True)
PIndexC.drop(PIndexC[(PIndexC['Publindex'] != "C")].index, inplace=True)
```

```python
Docentes=Nombre.value_counts("                 AUTOR").reset_index().rename(
           columns={'index': 'Autor', 0: 'count'})
Q1=Cuartil1.value_counts("                 AUTOR").reset_index().rename(
           columns={'index': 'Autor', 0: 'count'})
Q2=Cuartil2.value_counts("                 AUTOR").reset_index().rename(
           columns={'index': 'Autor', 0: 'count'})
Q3=Cuartil3.value_counts("                 AUTOR").reset_index().rename(
           columns={'index': 'Autor', 0: 'count'})
Q4=Cuartil4.value_counts("                 AUTOR").reset_index().rename(
           columns={'index': 'Autor', 0: 'count'})
A1=PIndexA1.value_counts("                 AUTOR").reset_index().rename(
           columns={'index': 'Autor', 0: 'count'})
A2=PIndexA2.value_counts("                 AUTOR").reset_index().rename(
           columns={'index': 'Autor', 0: 'count'})
B=PIndexB.value_counts("                 AUTOR").reset_index().rename(
           columns={'index': 'Autor', 0: 'count'})
C=PIndexC.value_counts("                 AUTOR").reset_index().rename(
           columns={'index': 'Autor', 0: 'count'})
```

```python
Clasificaciones = pd.DataFrame(columns=['Q1','Q2','Q3','Q4','A1','A2','B','C','Nacional','Internacional'],index = range(Docentes.shape[0]))
```

```python
CuadrosMaestros = pd.concat([Docentes,Clasificaciones], axis=1)
CuadrosMaestros = CuadrosMaestros.fillna('0')
```

```python
for l in range(len(CuadrosMaestros['                 AUTOR'])):
    for k in range(len(Q1['                 AUTOR'])):
        if CuadrosMaestros['                 AUTOR'][l] in Q1['                 AUTOR'][k]:
            CuadrosMaestros['Q1'][l] = Q1['count'][k]

for l in range(len(CuadrosMaestros['                 AUTOR'])):
    for k in range(len(Q2['                 AUTOR'])):
        if CuadrosMaestros['                 AUTOR'][l] in Q2['                 AUTOR'][k]:
            CuadrosMaestros['Q2'][l] = Q2['count'][k]    
            
for l in range(len(CuadrosMaestros['                 AUTOR'])):
    for k in range(len(Q3['                 AUTOR'])):
        if CuadrosMaestros['                 AUTOR'][l] in Q3['                 AUTOR'][k]:
            CuadrosMaestros['Q3'][l] = Q3['count'][k]                    

for l in range(len(CuadrosMaestros['                 AUTOR'])):
    for k in range(len(Q4['                 AUTOR'])):
        if CuadrosMaestros['                 AUTOR'][l] in Q4['                 AUTOR'][k]:
            CuadrosMaestros['Q4'][l] = Q4['count'][k]         

for l in range(len(CuadrosMaestros['                 AUTOR'])):
    for k in range(len(A1['                 AUTOR'])):
        if CuadrosMaestros['                 AUTOR'][l] in A1['                 AUTOR'][k]:
            CuadrosMaestros['A1'][l] = A1['count'][k]     

for l in range(len(CuadrosMaestros['                 AUTOR'])):
    for k in range(len(A2['                 AUTOR'])):
        if CuadrosMaestros['                 AUTOR'][l] in A2['                 AUTOR'][k]:
            CuadrosMaestros['A2'][l] = A2['count'][k]        

for l in range(len(CuadrosMaestros['                 AUTOR'])):
    for k in range(len(B['                 AUTOR'])):
        if CuadrosMaestros['                 AUTOR'][l] in B['                 AUTOR'][k]:
            CuadrosMaestros['B'][l] = B['count'][k]        
            
for l in range(len(CuadrosMaestros['                 AUTOR'])):
    for k in range(len(C['                 AUTOR'])):
        if CuadrosMaestros['                 AUTOR'][l] in C['                 AUTOR'][k]:
            CuadrosMaestros['C'][l] = C['count'][k]                              
```

```python
CuadrosMaestros=CuadrosMaestros.astype({'A1':'int','A2':'int','B':'int','C':'int','Q1':'int','Q2':'int','Q3':'int','Q4':'int'})
```

```python
CuadrosMaestros['Nacional'] = CuadrosMaestros['A1'] +  CuadrosMaestros['A2'] + CuadrosMaestros['B'] + CuadrosMaestros['C'] 
CuadrosMaestros['Internacional'] = CuadrosMaestros['Q1'] +  CuadrosMaestros['Q2'] + CuadrosMaestros['Q3'] + CuadrosMaestros['Q4'] 
CuadrosMaestros.to_excel('CuadrosMaestros.xlsx')
```
